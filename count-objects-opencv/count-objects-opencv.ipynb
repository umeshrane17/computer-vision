```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread("dataset/coins2.png")
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap="gray")
```

```python
blur = cv2.GaussianBlur(gray, (11,11), 0)
plt.imshow(blur, cmap="gray")
```

```python
canny = cv2.Canny(blur, 30, 150, 3)
plt.imshow(canny, cmap="gray")
```

```python
dilated = cv2.dilate(canny, (1,1), iterations=0)
plt.imshow(dilated, cmap="gray")
```

```python
(contours, hierarchy) = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, contours, -1, (0, 255, 0), 2)
plt.imshow(rgb)
```

```python
print("coins in the image : ", len(contours))
```
